**5. Extracting year, month and time of launch as separate columns**

```python
import pandas as pd

def extract_date_time(date_string):
  try:

    parts = date_string.split(',')[0].split(' ')[1:] + date_string.split(',')[1].strip().split(' ')[:2]

    date_time_string = ' '.join(parts)

    return date_time_string
  except IndexError:

    return None

df2['Extracted_Date_Time'] = df2['Datum'].apply(extract_date_time)
df2['Extracted_Date'] = df2['Extracted_Date_Time'].str.split(' ').str[0:2].str.join(' ')
df2['Extracted_Year'] = df2['Extracted_Date_Time'].str.split(' ').str[2]

df2.head()
```

```python
#Test
df2_test['Extracted_Date_Time'] = df2_test['Datum'].apply(extract_date_time)
df2_test['Extracted_Date'] = df2_test['Extracted_Date_Time'].str.split(' ').str[0:2].str.join(' ')
df2_test['Extracted_Year'] = df2_test['Extracted_Date_Time'].str.split(' ').str[2]
```

```python

month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

df2['Extracted_Month'] = df2['Extracted_Date_Time'].str.split(' ').str[0].map(month_mapping)
df2.head()
```

```python
#Test
df2_test['Extracted_Month'] = df2_test['Extracted_Date_Time'].str.split(' ').str[0].map(month_mapping)
```

```python
def extract_time_as_float(time_str):

    try:

        parts = time_str.split()
        time_part = parts[-1]
        hour, minute = map(int, time_part.split(':'))
        time_float =np.round(hour + (minute / 60),2)

        return time_float
    except (IndexError, ValueError):

        return None


time_str = "May 17 1967 16:05"
result = extract_time_as_float(time_str)
print(result)
```

**15. Combining it all into a pipeline**

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler,MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python

categorical_columns = ['Company Name', 'Rocket_type_Grouped','Country']
numerical_columns = [ 'Extracted_Year', 'Extracted_Month']

all_categories = {
    col: list(set(df4[col].unique()).union(set(df4_test[col].unique())))
    for col in categorical_columns
}


column_transformer = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(sparse_output=False, categories=[all_categories[col] for col in categorical_columns]), categorical_columns),  # Set explicit categories
        ('num', MinMaxScaler(), numerical_columns)
    ],
    remainder='passthrough'
)

pipeline = Pipeline(steps=[
    ('preprocessor', column_transformer)
])

train_transformed = pipeline.fit_transform(df4)
```

```python
#Test
test_transformed = pipeline.transform(df4_test)
```

```python
one_hot_columns = pipeline.named_steps['preprocessor'].transformers_[0][1].get_feature_names_out(categorical_columns)

final_column_names = list(one_hot_columns) + numerical_columns + ['Status Rocket']

train_transformed_df = pd.DataFrame(train_transformed, columns=final_column_names)
```
